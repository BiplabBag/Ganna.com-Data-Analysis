# 2. Data Preprocessing

## 2.1 Dealing With Duration

### for numerical operation we have to convert duration into seconds

```python
df['duration']
```

```python
for duration in df['duration'][:10]:
    print(duration, type(duration))
```

```python
#converting duration in seconds
durations_list = []
for duration in df['duration']:
        m = int(duration.split(':')[0])
        s = int(duration.split(':')[1])
        sec = m*60 + s 
        durations_list.append(sec)
print(durations_list[:5])        
```

```python
df['duration'] = durations_list    #putting duration in seconds in df
```

## 3.5 Average duration of songs in each langauge

```python
for lang in df['language'].unique():
        count = 0
        total_duration = 0
        for duration in df[df['language'] == lang]['duration']:
                count += 1 
                total_duration += duration
        print('average duraion of', lang ,'song is',total_duration/count )        
```

## 3.8 Who sang highest no of songs

```python
dic = {}
# adding singers and their no of songs in dictionary
for singers in df['singer']:
    singer = singers.split('|')
    for i in singer:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] += 1
# from dictionary getting singer name who sang highest no of songs
highest_song = 0
singer = None
for key,value in dic.items():
    if value > highest_song:
        highest_song = value
        singer = key
print(singer,highest_song)
```

## 3.9 Who sang highest no of songs in each language

```python
for lang in df['language'].unique():
        dic = {}
        # adding singers and their no of songs in dictionary in particular language
        for singers in df[df['language'] == lang]['singer']:
            singer = singers.split('|')
            for i in singer:
                if i not in dic:
                    dic[i] = 1
                else:
                    dic[i] += 1
        # from dictionary getting singer name who sang highest no of songs in perticular language
        highest_song = 0
        singer = None
        for key,value in dic.items():
            if value > highest_song:
                highest_song = value
                singer = key
        print(singer,'has highest', highest_song, 'songs in' , lang )
```
